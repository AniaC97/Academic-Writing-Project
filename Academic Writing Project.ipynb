## *Importing page and printing its content*

```python
import requests
page = requests.get("https://www.nytimes.com/2021/05/26/movies/cruella-review.html")
print(page)
```

```python
print(page.content)
```

## *Scraping the page content using the Beautiful Soup*

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
paragraphs = soup.find_all('p')
```

```python
for el in paragraphs:
    text = el.get_text().strip()
    if len(text) > 0:
        print(text)
```

## *Tokenization of words using NLTK*

The next step in the analysis is tokenization of words which occured in the text.

Tokenization is a key element of text analysis. During the process of tokenization, the text is separated into smaller pieces called tokens. The text may be tokenized by:

*	*word* – because of this type, the user can find words which were used most frequently in the text as well as identify to which part of speech they belong
*	*sentence* – this type of tokenization allows the user to analyse the whole sentence. It shows context and how words relate to one another.


```python
import nltk
```

```python
for el in paragraphs:
    text = el.get_text().strip()
    if len(text) > 0:
        tokens = nltk.word_tokenize(text)
        tagged = nltk.pos_tag(tokens)
        print(tagged)
```

After that, NLTK can be used to count appearances of different parts of speech in each paragraph.

```python
for el in paragraphs:
    text = el.get_text().strip()
    if len(text) > 0:
        tokens = nltk.word_tokenize(text)
        tagged = nltk.pos_tag(tokens)
        
        from collections import Counter
        counts = Counter(tag for word,tag in tagged)
        print(counts)
```

## *Frequency Distribution*

In this analysis frequency distribution is used to record words which were used most frequently in the New York Times review.

```python
stripped_text = soup.get_text(separator=" ")
```

```python
import re

def remove_url(text_data):
    return re.sub(r"http\S+", "", text_data)

text2 = remove_url(stripped_text)
```

### Stop Words

As the next step of the analysis English stop words were deleted from the text.

Stop words is a set of words which are commonly used in a given language. As stop words determiners, coordinating conjunctions and prepositions might be considered. The list of stop words depends on the theme of the analysis as well.

```python
english_stop_words = nltk.corpus.stopwords.words('english')
print(english_stop_words)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(text2)

tokens = [token.lower() for token in tokens]

stops = []
for word in tokens:
    if word not in english_stop_words:
        stops.append(word)
        
print(stops)
```

```python
final_words = [word for word in stops if word.isalpha()]
for word in stops:
        final_words.append(word)
        
print(final_words)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()

freqdist1 = nltk.FreqDist(final_words)
freqdist1.plot(20)
```
